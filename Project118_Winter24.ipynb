```python
##Part C

#race vs. desire to graduate from a 4-year university program in Core_df

def black_graduate_4year(row):
    '''This function filters out black students who hope to graduate from a 4-year university program'''
    
    return (row['V2151'] == 1) & (row['V2188'] == 1)

def white_graduate_4year(row):
    '''This function filters out white students who hope to graduate from a 4-year university program'''
    
    return (row['V2151'] == 2) & (row['V2188'] == 1)

def hispanic_graduate_4year(row):
    '''This function filters out hispanic students who hope to graduate from a 4-year university program'''
    
    return (row['V2151'] == 3) & (row['V2188'] == 1)

# Adding to original core dataframe and summing

black_graduate_sum = Core_df.apply(black_graduate_4year, axis=1).sum()
white_graduate_sum = Core_df.apply(white_graduate_4year, axis=1).sum()
hispanic_graduate_sum = Core_df.apply(hispanic_graduate_4year, axis=1).sum()

#percentages

black_percent_graduate = black_graduate_sum/934
white_percent_graduate = white_graduate_sum/4090
hispanic_percent_graduate = hispanic_graduate_sum/1874

desire_to_graduate_df = pd.DataFrame({
    'Racial Group': ['Black', 'White', 'Hispanic'],
    'Percentage of Students Desiring to Graduate': [black_percent_graduate, white_percent_graduate, hispanic_percent_graduate]
})

plt.savefig("Table2.png", bbox_inches='tight')

desire_to_graduate_df
```

```python
#Create the bar plot
plt.figure(figsize=(8, 6))
plt.bar(desire_to_graduate_df['Racial Group'], desire_to_graduate_df['Percentage of Students Desiring to Graduate'], color=['skyblue', 'lightgreen', 'salmon'])

plt.ylabel('Percent', fontsize=14)
plt.title('Figure 4: Desire to Graduate from a 4-Year University Program by Racial Group', fontsize=14)
plt.xticks(fontsize=11)

plt.savefig("Figure4.png")

plt.show()
```

```python
#sex vs. political leanings in Core_df

def male_conservative(row):
    '''This function filters out male students who self-describe their political beliefs as very conservative'''
    
    return (row['V2150'] == 1) & (row['V2167'] == 1 or row['V2167'] == 2)

def female_conservative(row):
    '''This function filters out female students who self-describe their political beliefs as very conservative'''
    
    return (row['V2150'] == 2) & (row['V2167'] == 1 or row['V2167'] == 2)

def male_liberal(row):
    '''This function filters out male students who self-describe their political beliefs as very liberal'''
    
    return (row['V2150'] == 1) & (row['V2167'] == 4 or row['V2167'] == 5)
    
def female_liberal(row):
    '''This function filters out female students who self-describe their political beliefs as very liberal'''
    
    return (row['V2150'] == 2) & (row['V2167'] == 4 or row['V2167'] == 5)

# Applying the filtering functions
male_conservative_sum = Core_df.apply(male_conservative, axis=1).sum()
female_conservative_sum = Core_df.apply(female_conservative, axis=1).sum()
male_liberal_sum = Core_df.apply(male_liberal, axis=1).sum()
female_liberal_sum = Core_df.apply(female_liberal, axis=1).sum()

#percantages

male_conservative_percentage = male_conservative_sum/4284
female_conservative_percentage = female_conservative_sum/4191

male_liberal_percentage = male_liberal_sum/4284
female_liberal_percentage = female_liberal_sum/4191

sex_political_leaning_df = pd.DataFrame({
    'Gender': ['Male', 'Female'],
    'Conservatives': [male_conservative_percentage, female_conservative_percentage],
    'Liberals': [male_liberal_percentage, female_liberal_percentage]
})

sex_political_leaning_df
```

```python
genders = ['Male', 'Female']
conservative_percentages = [male_conservative_percentage, female_conservative_percentage]
liberal_percentages = [male_liberal_percentage, female_liberal_percentage]

#looked up a lot of this formatting in order to make it look sleek

bar_width = 0.35

r1 = range(len(genders))
r2 = [x + bar_width for x in r1]

plt.bar(r1, conservative_percentages, color='blue', width=bar_width, edgecolor='grey', label='Conservative')
plt.bar(r2, liberal_percentages, color='orange', width=bar_width, edgecolor='grey', label='Liberal')

plt.xlabel('Sex')
plt.ylabel('Percentage')
plt.xticks([r + bar_width/2 for r in range(len(genders))], genders)
plt.title('Figure 5: Political Leaning by Sex')

plt.legend(loc = 'upper center')

plt.savefig("Figure5.png")

plt.tight_layout()
plt.show()
```

```python
#urbanicity vs. religion

def MSA_not_important(row):
    '''This function filters out urban-dwelling students who self-describe the role of religion 
    in their life as "not important"'''
    
    return (row['V17'] == 1) & (row['V2170'] == 1)

def MSA_a_little_important(row):
    '''This function filters out urban-dwelling students who self-describe the role of religion 
    in their life as "a little important"'''
    
    return (row['V17'] == 1) & (row['V2170'] == 2)

def MSA_pretty_important(row):
    '''This function filters out urban-dwelling students who self-describe the role of religion 
    in their life as "pretty important"'''
    
    return (row['V17'] == 1) & (row['V2170'] == 3)

def MSA_very_important(row):
    '''This function filters out urban-dwelling students who self-describe the role of religion 
    in their life as "very important"'''
    
    return (row['V17'] == 1) & (row['V2170'] == 4)

def noMSA_not_important(row):
    '''This function filters out non-urban-dwelling students who self-describe the role of religion 
    in their life as "not important"'''
    
    return (row['V17'] == 0) & (row['V2170'] == 1)

def noMSA_a_little_important(row):
    '''This function filters out non-urban-dwelling students who self-describe the role of religion 
    in their life as "a little important"'''
    
    return (row['V17'] == 0) & (row['V2170'] == 2)

def noMSA_pretty_important(row):
    '''This function filters out non-urban-dwelling students who self-describe the role of religion 
    in their life as "pretty important"'''
    
    return (row['V17'] == 0) & (row['V2170'] == 3)

def noMSA_very_important(row):
    '''This function filters out non-urban-dwelling students who self-describe the role of religion 
    in their life as "very important"'''
    
    return (row['V17'] == 0) & (row['V2170'] == 4)

#adding to original core dataframe and summing

MSA_not_important_sum = Core_df.apply(MSA_not_important, axis=1).sum()
MSA_a_little_important_sum = Core_df.apply(MSA_a_little_important, axis=1).sum()
MSA_pretty_important_sum = Core_df.apply(MSA_pretty_important, axis=1).sum()
MSA_very_important_sum = Core_df.apply(MSA_very_important, axis=1).sum()

noMSA_not_important_sum = Core_df.apply(noMSA_not_important, axis=1).sum()
noMSA_a_little_important_sum = Core_df.apply(noMSA_a_little_important, axis=1).sum()
noMSA_pretty_important_sum = Core_df.apply(noMSA_pretty_important, axis=1).sum()
noMSA_very_important_sum = Core_df.apply(noMSA_very_important, axis=1).sum()

#percents

MSA_not_important_percent = MSA_not_important_sum/7853
MSA_a_little_important_percent = MSA_a_little_important_sum/7853
MSA_pretty_important_percent = MSA_pretty_important_sum/7853
MSA_very_important_percent = MSA_very_important_sum/7853

noMSA_not_important_percent = noMSA_not_important_sum/1746
noMSA_a_little_important_percent = noMSA_a_little_important_sum/1746
noMSA_pretty_important_percent = noMSA_pretty_important_sum/1746
noMSA_very_important_percent = noMSA_very_important_sum/1746


MSA_religion_importance_df = pd.DataFrame({
    'Importance Level': ['Not Important', 'A Little Important', 'Pretty Important', 'Very Important'],
    'Percent': [MSA_not_important_percent, MSA_a_little_important_percent, MSA_pretty_important_percent, MSA_very_important_percent]
})

noMSA_religion_importance_df = pd.DataFrame({
    'Importance Level': ['Not Important', 'A Little Important', 'Pretty Important', 'Very Important'],
    'Percent': [noMSA_not_important_percent, noMSA_a_little_important_percent, noMSA_pretty_important_percent, noMSA_very_important_percent]
})

MSA_religion_importance_df
```
